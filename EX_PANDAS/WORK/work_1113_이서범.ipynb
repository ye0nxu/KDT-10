```python
import numpy as np
import pandas as pd
```

### 시리즈 생성 <hr>

```python
## 시리즈 생성
sr1 = pd.Series([100,200,300],name='점수')
sr1
```

```python
## 시리즈 추가 -> O
sr2 = pd.Series(index=['철수','영희','아름'],data=[100,200,300])
sr2
```

```python
## 시리즈의 생성 (딕셔너리) -> X
sr3 = pd.Series({
    '철수':100,
    '영희':200,
    '아름':300
    })
sr3
```

```python
## 시리즈 생성 -> X
index = ['삼성전자','셀트리온','카카오','삼성전자우','현대바이오']
data = [73000,356000,367000,68600,34150]
sr4 = pd.Series(data=data, index=index)
sr4
```

```python
## 인덱스 없이 시리즈에 저장된 데이터만을 출력해보세요.
data = [100, 200, 300]
index = ["철수", "영희", "아름"]
s = pd.Series(data, index)
s.index
```

```python
## 시리즈를 딕셔너리로 변경해 보세요.
data = [100, 200, 300]
index = ["철수", "영희", "아름"]
s = pd.Series(data, index)
s = s.to_dict()
s
```

```python
## 시리즈를 엑셀 파일(data.xlsx)에 저장해 보세요.
data = [100, 200, 300]
index = ["철수", "영희", "아름"]
s = pd.Series(data, index)
s.to_excel('./data.xlsx')
```

```python
## 엑셀 파일(data.xlsx)을 읽어 오세요. -> X
# 데이터를 불러오면 DF인데, 해당 컬럼을 인덱싱하면 series를 얻을 수 있음. 
pd.read_excel('./data.xlsx',index_col=0)[0]
```

## 시리즈 데이터 타입 <hr>

```python
## 시리즈의 데이터 타입을 출력해보세요. -> X
# dtype는 ()강 없음. 메소드가 아니라 ''임
data = [100, 200, 300]
index = ["철수", "영희", "아름"]
s = pd.Series(data, index)

s.dtype
```

```python
## 다음 시리즈의 데이터 타입은 무엇일까요? -> O
data = [  10,  "가" ]
s = pd.Series( data )
s.dtype
# 정답 : object
```

```python
## 시리즈에서 인덱스의 데이터 타입을 출력해보세요. -> O
s.index.dtype
```

```python
## 시리즈 객체로 생성한 후 시리즈의 name 속성에 '종가'를 
# 인덱스의 name 속성에 날짜를 설정해보세요.
index = ['2020-12-11','2020-12-10','2020-12-09','2020-12-08','2020-12-07']
data = [73400,72900,73900,71700,72900]
s=pd.Series(index=index, data=data)
s.name='종가'
s.index.name='날짜'
s
```

```python
## 인덱스 이름을 icecream으로 변경 -> O
data = {"apple": 1500, "banana": 3000}
s = pd.Series(data)
s.index.name='icecream'
s
```

```python
## 다음 시리즈 객체의 이름을 'number'로 지정하세요.
s = pd.Series(data=[1, 2, 3])
s.name = 'number'
```
