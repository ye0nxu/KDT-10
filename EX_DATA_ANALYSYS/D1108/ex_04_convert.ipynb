```python
##
import pandas as pd
```

```python
pd.to_datetime("2025-11-08")
pd.to_datetime(["2025-11-08", "2025-11-09"])

# ISO 형태
pd.to_datetime("2025-11-08 14:30:05", format="%Y-%m-%d %H:%M:%S")
# 한글 리터럴 포함
pd.to_datetime("2023년 12월 25일", format="%Y년 %m월 %d일")
pd.to_datetime()
```

```python
## => Period는 구간이므로, 시각(10:20:30)은 사라지고
## => “그 날/그 달/그 분기” 같은 라벨이 됨.
t = pd.Timestamp('2025-03-15 10:20:30')

print( t.to_period('D') ) # 일 단위 → Period('2025-03-15', 'D')
print( t.to_period('M') ) # 월 단위 → Period('2025-03', 'M')
print( t.to_period('Q') ) # 분기 단위 → Period('2025Q1', 'Q-DEC')
print( t.to_period('Y') ) # 연 단위 → Period('2025', 'A-DEC')
```
