```python
## 모듈 로딩
import pandas as pd
```

```python
## read_csv() 함수로 파일 읽어와서 df로 변환
df = pd.read_csv('./data/stock-data.csv')
df2 = df.copy()

## 문자열인 날짜 데이터를 판다스 Timestamp로 변환
df2['new_Date'] = pd.to_datetime(df2['Date'])   #df에 새로운 열로 추가
df2.head()
df2.info()
```

```python
## dt 속성을 이용하여 new_Date 열의 년월일 정보를 년, 월, 일로 구분
df2['Year']  = df2['new_Date'].dt.year
df2['Month'] = df2['new_Date'].dt.month
df2['Day']   = df2['new_Date'].dt.day
df2.head()
```

```python
# Timestamp를 Period로 변환하여 년월일 표기 변경
df3 = df.copy()
df3['new_Date'] = pd.to_datetime(df2['Date']) 
df3['Date_yr'] = df3['new_Date'].dt.to_period(freq='Y')
df3['Date_m']  = df3['new_Date'].dt.to_period(freq='M')
df3.head()
```

```python
# 2018년 6월 데이터만 추출 (문자열 속성 활용)
df_june = df3[df3['Date_m'].astype(str).str.fullmatch('2018-06')]
df_june.head()
```

```python
# 2018년 6월 데이터만 추출 (Period 객체 활용)
df_june2 = df3[df3['Date_m'] == pd.Period('2018-06')]
df_june2.head()
```

```python
# 원하는 열을 새로운 행 인덱스로 지정
df_june = df_june.set_index('new_Date')
df_june.head()
```
