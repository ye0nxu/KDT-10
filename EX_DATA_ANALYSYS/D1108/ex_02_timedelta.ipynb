```python
##
import pandas as pd
```

```python
##=======================================
## Timedelta 인스턴스 생성 및 속성들
##========================================
td1=pd.Timedelta(1, 'd')
td2=pd.Timedelta(days=1)

print(f'td1=>{td1} {type(td1)}')
print(f'td2=>{td2} {type(td2)}')

## 속성들:인스턴스변수명.속성명
print(f'.days : {td2.days}')
print(f'.seconds : {td2.seconds}')
print(f'.total_seconds() : {td2.total_seconds()}')
print(f'.components : {td2.components}')
```

```python
## ==========================================

## Timestamp & Timedelta 연산

## ==========================================

ts = pd.Timestamp('2025-11-08 20:12:12')

print(f'ts => {ts}')

## 산술 연산

ret = ts + pd.Timedelta('2h') 

print(f'ts+2H 즉, 2시간 뒤 ret => {ret} / {type(ret)}')

ret = ts - pd.Timedelta(1, 'd') 

print(f'ts-1d 즉, 하루 전 ret => {ret}  / {type(ret)}')

ret = ts + pd.Timedelta(1, 'd') 

print(f'ts+1d 즉, 하루 뒤 ret => {ret}  / {type(ret)}')
```

```python

## ==========================================

## Timestamp & Timestamp 연산

## ==========================================

ts1 = pd.Timestamp('2025-11-08 20:12:12')

ts2 = pd.Timestamp('2024-11-08 20:12:12')

## 산술 연산

ret = ts1 - ts2

print(f'ts1 - ts2 ret => {ret} / {type(ret)}')

## => 미지원 기능 : unsupported operand type(s) for +: 'Timestamp' and 'Timestamp'

# ret = ts1 + ts2

# print(f'ts1 + ts2 ret => {ret} / {type(ret)}')

## => 뺄셈 제외한 나머지 연산들은 Timedelta 객체와 진행함!
```

```python
## ==========================================

## Timestamp & 메서드들

## ==========================================

ts1 = pd.Timestamp('2025-11-08 20:15:12')

print(f"floor(h)     : {ts1.floor('h')}")

print(f"ceil(h)      : {ts1.ceil('h')}\n")

print(f"floor(D)     : {ts1.floor('D')}")

print(f"ceil(D)      : {ts1.ceil('D')}\n")

print(f"round(10min) : {ts1.round('10min')}")

print(f"ceil(min)    : {ts1.ceil('min')}")
```
