```python
##
import pandas as pd
```

```python
##=======================================
## Timestamp 인스턴스 생성 및 속성들
##========================================
ts1=pd.Timestamp('2025-11-08 17:25')
ts2=pd.Timestamp(year=2030,month=9,day=12)

print(f'ts1=>{ts1} {type(ts1)}')
print(f'ts2=>{ts2} {type(ts2)}')

## 속성들:인스턴스변수명.속성명
print(f'ts1.year:{ts1.year}')
print(f'ts1.dayL{ts1.day}/{ts1.day_name()}')

print(f'ts2.year:{ts2.year}')
print(f'ts2.dayL{ts2.day}/{ts2.day_name()}')
print(f'ts2.month{ts2.month}/{ts2.month_name()}')
```

```python
## 현재 지역의 날짜 시간 정보
now_= ts2.now()
print(f"현재 지역 날짜 시간:{now_}")
```

```python
## Index들 ...

## [1] RangeIndex : pandas에서 위치지정한 값으로 구성된 인덱스
sr = pd.Series([11,22,33,44,55])

rangeIdx = sr.index
print(f'rangeIdx => {rangeIdx}')

## [2] 라벨인덱스 / 인덱스 종류에 따라 다양
sr = pd.Series([11,22,33,44,55], index=[['A','B','C','D','E']])

rangeIdx = sr.index
print(f'rangeIdx => {rangeIdx}')
```
