#### [ PYTHON & MYSQL 연동 ]


```python
## 모듈 로딩
import pymysql

print(f'pymysql v.{pymysql.__version__}')
```

```python
## RDBMS 정보
SEVER_IP = '172.30.1.33'
USER_ID = 'user2'
USER_PW = 'user2'
DB_NAME = 'ems'
CHARSET = 'utf8mb4'
```

```python
## =============================================================
## DB 연결
## =============================================================
conn = pymysql.connect(host=SEVER_IP,
                       user=USER_ID,
                       password=USER_PW,
                       database=DB_NAME,
                       cursorclass=pymysql.cursors.DictCursor)
```

```python
## =============================================================
## DB - SELECT 처리
## =============================================================
ALL_DATA_SQL = 'SELECT * FROM user;'

with conn.cursor() as cursor:
    ## 모든 데이터 불러오기
    cursor.execute(ALL_DATA_SQL)
    result = cursor.fetchall()
    print(result)
```

```python
## =============================================================
## DB - INSERT 처리
## =============================================================
INSERT_DATA_SQL = "INSERT INTO user VALUES (11111, '베트맨', 23, '부산', '부산시', '여')"


with conn.cursor() as cursor:
    cursor.execute(INSERT_DATA_SQL)
    conn.commit()
```

```python
## =============================================================
## DB - DELETE 처리
## =============================================================
DELETE_DATA_SQL = "DELETE FROM user WHERE id=%s;"


with conn.cursor() as cursor:
    cursor.execute(DELETE_DATA_SQL, (11111,))
    conn.commit()
```

```python
## =============================================================
## DB - UPDATE 처리
## =============================================================
UPDATE_DATA_SQL = "UPDATE user SET name=%s WHERE id=%s;"


with conn.cursor() as cursor:
    cursor.execute(UPDATE_DATA_SQL, ('원더우먼',10))
    conn.commit()
```

```python
## =============================================================
## DB 해제
## =============================================================
conn.close()
```
