### 【 Matplotlib 자원 설정 】
- rc : resource configuration 시각화에 필요한 다양한 설정값들 저장
- 한글 폰트 미지원 / 마이너스(-) 기호 깨져서 출력

[1] 모듈 로딩 및 기본 설정 확인 <hr>

```python
## 모듈 로딩
from matplotlib import rcParams
import pandas as pd
```

```python
## 현재 설정된 값 들 출력
print(rcParams)
```

```python
## 현재 설정된 특정 설정값들 출력
## - axis
for key in rcParams.keys():
    if 'axis' in key:
        print(key)

## - font
print()
for key in rcParams.keys():
    if 'font' in key:
        print(key)
```

[2] 마이너스 기호 설정 <hr>
- 기본값 : 유니코드 '-'(U+2212)
- 문제점 : 한글 폰트에 해당 코드값 없음 => □(깨짐)으로 출력

```python
## [설정방법 1]
from matplotlib import rcParams

## Windows 유니코드 '-'(U+2212) 대신 => ASCII '-' 사용 설정
rcParams['axes.unicode_minus'] = False
```

```python
## [설정방법 2]
import matplotlib.pyplot as plt

plt.rcParams['axes.unicode_minus'] = False
```
