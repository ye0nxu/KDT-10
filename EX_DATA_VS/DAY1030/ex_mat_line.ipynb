### [ 선 그래프 ]
- 시간의 흐름에 따른 추세/변화/트렌드 표현


[1] 모듈 로딩 및 데이터 준비 <hr>



```python
## 모듈 로딩
import pandas as pd                  ## 데이터 분석 관련 모듈
import matplotlib.pyplot as plt      ## 그래프 그리는 그래프 관련 모듈
import koreanize_matplotlib          ## 한글이 들어가야하니깐 한글 폰트 관련 모듈
```

```python
## 데이터 준비
data = {'A' : [10,20,30,40,50], 'B':[15,25,35,45,55]}

dataDF=pd.DataFrame(data)
display(dataDF)
```

[2] A에 따른 B에 변화 시각화 <hr>

```python
## ========================================
# ## 선 그래프 함수 : plt.plot()
## ========================================
plt.plot(dataDF['A'], dataDF['B'])

## 그래프 기본 설정
plt.xlabel('A')
plt.ylabel('B')
plt.title('[A에 따른 B의 변화]')


## 그래프 화면 출력 => 이거 안해주면 그래프 설명에 대한 메세지 나옴
plt.show()
```

```python
## ========================================
# ## 선 그래프 함수 : plt.plot(y, fmt)
## ========================================
plt.plot(dataDF['A'], 'r-')         ## 빨간색 실선
plt.plot(dataDF['A']+10, 'b--')     ## 파랑색 점선
plt.plot(dataDF['A']-10, 'g^')      ## 녹색 삼각형마크 선X 
plt.plot(dataDF['A']-5, 'm^-')      ## mageta 삼각형마크 실선

## 그래프 기본 설정
plt.xlabel('A')
plt.title('[A 변화]')


## 그래프 화면 출력 => 이거 안해주면 그래프 설명에 대한 메세지 나옴
plt.show()
```

```python
## ========================================
# ## 선 그래프 함수 : plt.plot(y, fmt)
## ========================================
plt.plot(dataDF['A'], 'r-', label='A')         ## 빨간색 실선
plt.plot(dataDF['A']+10, 'b--', label='A+10')     ## 파랑색 점선
plt.plot(dataDF['A']-10, 'g^', label='A-10')      ## 녹색 삼각형마크 선X 
plt.plot(dataDF['A']-5, 'm^-', label='A-5')      ## mageta 삼각형마크 실선

## 그래프 기본 설정
plt.xlabel('A')
plt.title('[A 변화]')
plt.legend()                     ## 범례출력 -> 라벨쓰고 라벨 출력할려면 이거 써야함

## 그래프 화면 출력 => 이거 안해주면 그래프 설명에 대한 메세지 나옴
plt.show()
```
