### 【 데이터 시각화 - Area Graph 】
- 선 그래프에서 추가된 기능
- 특정 영역을 강조/집중해서 표현하고 싶은 경우 사용 
- 너무 많은 조각은 오히려 가독성 떨어지고 집중 X
- 함수 : matplotlib.pyplot.fill_between()/fill_betweenx()

```python
### [1] 모듈 로딩 
import matplotlib.pyplot as plt
import koreanize_matplotlib 
import pandas as pd
```

```python
## ==================================================
## [2] 선 그래프 
## ==================================================
## - 데이터 
xData = [1, 2, 3, 4]
yData = [2, 3, 5,10]

## - 선 그래프 인스턴스 생성 : 기본값으로 선 그래프 생성
line = plt.plot(xData, yData )

## - 그래프 기본 요소 설정
plt.title("기본 선그래프")
plt.xlabel("X 데이터")
plt.ylabel("Y 데이터")
plt.grid()              ## 눈금선 설정

## - 화면에 그래프 출력
plt.show()    
```

```python
## ==================================================
## [2] 선 그래프 + 영역 설정 
## ==================================================
## - 데이터 
##   idx 0  1  2  3
xData = [1, 2, 3, 4]
yData = [2, 3, 5,10]

## - 선 그래프 인스턴스 생성 : 기본값으로 선 그래프 생성
line = plt.plot(xData, yData )

## - 특정 영역 선아래 채우기 : xData[1:3] ==> 2, 3, yData[1:3] ==> 3, 5
##   alpha 매개변수: 투명도 설정. 1 - 불투명, 0 - 투명
plt.fill_between(xData[1:3], yData[1:3], alpha=0.2)

## - 그래프 기본 요소 설정
plt.title("기본 선그래프")
plt.xlabel("X 데이터")
plt.ylabel("Y 데이터")
plt.grid()              ## 눈금선 설정

## - 화면에 그래프 출력
plt.show()  
```

```python
## ==================================================
## [2] 선 그래프 + 영역 설정 
## ==================================================
## - 데이터 
##   idx 0  1  2  3
xData = [1, 2, 3, 4]
yData = [2, 3, 5,10]

## - 선 그래프 인스턴스 생성 : 기본값으로 선 그래프 생성
line = plt.plot(xData, yData )

## - 특정 영역 선 아래 채우기 : xData[1:3] ==> 2, 3, yData[1:3] ==> 3, 5
##   alpha 매개변수: 투명도 설정. 1 - 불투명, 0 - 투명
plt.fill_between(xData[1:3],  yData[1:3], alpha=0.2)

## - 특정 영역 선 위 채우기
plt.fill_betweenx(yData[1:3], xData[1:3], alpha=0.2)

## - 그래프 기본 요소 설정
plt.title("기본 선그래프")
plt.xlabel("X 데이터")
plt.ylabel("Y 데이터")
plt.grid()              ## 눈금선 설정

## - 화면에 그래프 출력
plt.show()  
```

```python
## ==================================================
## [2] 선 그래프 + 영역 설정 
## ==================================================
## - 데이터 
##   idx 0  1  2  3
xData = [1, 2, 3, 4]
yData = [2, 3, 5,10]

## - 선 그래프 인스턴스 생성 : 기본값으로 선 그래프 생성
line = plt.plot(xData, yData )

## - 특정 영역 선위 채우기 : xData[2:4] ==> 3, 4, yData[2:4] ==> 5, 10
## - fill_betweenx(y, x) : y축 기준
##                 alpha 매개변수: 투명도 설정. 1 - 불투명, 0 - 투명
plt.fill_betweenx(yData[2:4], xData[2:4], alpha=0.2)

## - 그래프 기본 요소 설정
plt.title("기본 선그래프")
plt.xlabel("X 데이터")
plt.ylabel("Y 데이터")
plt.grid()              ## 눈금선 설정

## - 화면에 그래프 출력
plt.show()  
```
