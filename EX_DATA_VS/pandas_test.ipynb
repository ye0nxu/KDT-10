```python
df_m = pd.read_excel('./DATA/매매_41133_202101.xlsx', index_col=0)
df_m.head()
```

```python
## 거래금액 컬럼에는 문자열(object) 타입으로 데이터가 저장돼 있습니다. -> O
# 콤마를 제거하고 데이터 타입을 숫자로 변경하세요.
df_m.거래금액 = df_m.거래금액.str.replace(',','').astype('int')
df_m.info()
df_m.head()
```

```python
## 다양한 크기의 아파트 가격을 비교하기위해 매매평단가를 계산합니다. 
#  일반적으로 평단가는 거래금액과 공급면적으로 계산하지만, 
#  이번문제에서는 주어진 데이터 거래금액과 전용면적으로 매매평단가를 계산해 봅시다.
#  m^2 를 평으로 환산할 때는 3.3으로 값을 나눕니다.
df_m['매매평단가'] = df_m['거래금액'] / (df_m['전용면적'] / 3.3) 
df_m.head()
```

```python
## 주어진 성남시 데이터에는 총 8개의 법정동이 존재합니다. 동별로 매매평단가의 평균을 계산하세요.
df1 = df_m[['법정동','매매평단가']].groupby(by=['법정동'], observed=True).mean(numeric_only=True)
df1
```

```python
# 하대원동의 아파트별 매매평단가와 매매건수를 출력하세요.
df_m['매매건수'] = 1

dfk = df_m[df_m['법정동'] == '하대원동']
dfk[['매매평단가','아파트','매매건수']].groupby(by=['아파트'], observed=True).agg({'매매평단가':'mean', '매매건수':'count'})
```

```python
df_m.월 = df_m.월.astype('str')
df_m.일 = df_m.일.astype('str')
df_m.년 = df_m.년.astype('str')

def f(x):
    if len(x) == 1:
        return f"0{x}"
    else:
        return x

df_m['월'] = df_m['월'].map(f)
df_m['일'] = df_m['일'].map(f)

df_m['날짜'] = df_m['년'] + df_m['월'] + df_m['일'] 
df_m['날짜'] = pd.to_datetime(df_m['날짜'])

temp = df_m[ ["날짜", "법정동", "아파트", "매매평단가"] ]
temp.head(2)
```

```python
data = [
    ["2차전지(생산)", "SK이노베이션", 10.19, 1.29],
    ["해운", "팬오션", 21.23, 0.95],
    ["시스템반도체", "티엘아이", 35.97, 1.12],
    ["해운", "HMM", 21.52, 3.20],
    ["시스템반도체", "아이에이", 37.32, 3.55],
    ["2차전지(생산)", "LG화학", 83.06, 3.75]
]

columns = ["테마", "종목명", "PER", "PBR"]
df = pd.DataFrame(data=data, columns=columns)
display(df[['종목명','PER','PBR']].groupby(by=['종목명'],observed=True).size())
df.종목명.value_counts()
```
