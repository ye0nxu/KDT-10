```python
import pandas as pd
import matplotlib.pyplot as plt
import koreanize_matplotlib
```

```python
df = pd.read_csv('../DATA/iris.csv')
df.head()
```

```python
df = df.set_index('variety')
df.head()
```

```python
## 종류별로 sepal.length 추출
dfSL1 = df.loc['Setosa','sepal.length']
dfSL2 = df.loc['Versicolor','sepal.length']
dfSL3 = df.loc['Virginica','sepal.length']
# print(dfSL1)
# print(dfSL1.dtype)

## 종류별로 그리기
plt.plot(dfSL1.index, dfSL1.values)
plt.plot(dfSL2.index, dfSL2.values)
plt.plot(dfSL3.index, dfSL3.values)

plt.show()
```

```python
## 종류별로 sepal.width 그리기
dfSL1 = df.loc['Setosa','sepal.width']
dfSL2 = df.loc['Versicolor','sepal.width']
dfSL3 = df.loc['Virginica','sepal.width']
# print(dfSL1)
# print(dfSL1.dtype)


## 종류별로 그리기
plt.plot(dfSL1.index, dfSL1.values)
plt.plot(dfSL2.index, dfSL2.values)
plt.plot(dfSL3.index, dfSL3.values)

plt.show()
```

```python
## 종류별로 petal.length 추출
dfSL1 = df.loc['Setosa','petal.length']
dfSL2 = df.loc['Versicolor','petal.length']
dfSL3 = df.loc['Virginica','petal.length']
# print(dfSL1)
# print(dfSL1.dtype)

## 종류별로 그리기
plt.plot(dfSL1.index, dfSL1.values)
plt.plot(dfSL2.index, dfSL2.values)
plt.plot(dfSL3.index, dfSL3.values)

plt.show()
```

```python
## 종류별로 petal.width 추출
dfSL1 = df.loc['Setosa','petal.width']
dfSL2 = df.loc['Versicolor','petal.width']
dfSL3 = df.loc['Virginica','petal.width']
# print(dfSL1)
# print(dfSL1.dtype)

## 종류별로  그리기
plt.plot(dfSL1.index, dfSL1.values)
plt.plot(dfSL2.index, dfSL2.values)
plt.plot(dfSL3.index, dfSL3.values)

plt.show()
```

```python
## 모아서 그리기
fig, axes = plt.subplots(2,2,constrained_layout=True)

#################################################
## 종류별로 sepal.length 추출
dfSL1 = df.loc['Setosa','sepal.length']
dfSL2 = df.loc['Versicolor','sepal.length']
dfSL3 = df.loc['Virginica','sepal.length']
# print(dfSL1)
# print(dfSL1.dtype)

## 종류별로 그리기
axes[0,0].plot(dfSL1.index, dfSL1.values)
axes[0,0].plot(dfSL2.index, dfSL2.values)
axes[0,0].plot(dfSL3.index, dfSL3.values)

axes[0,0].set_title('sepal.length')

####################################################
## 종류별로 sepal.width 그리기
dfSL1 = df.loc['Setosa','sepal.width']
dfSL2 = df.loc['Versicolor','sepal.width']
dfSL3 = df.loc['Virginica','sepal.width']
# print(dfSL1)
# print(dfSL1.dtype)

## 종류별로 그리기
axes[0,1].plot(dfSL1.index, dfSL1.values)
axes[0,1].plot(dfSL2.index, dfSL2.values)
axes[0,1].plot(dfSL3.index, dfSL3.values)

axes[0,1].set_title('sepal.width')

#####################################################
## 종류별로 petal.length 추출
dfSL1 = df.loc['Setosa','petal.length']
dfSL2 = df.loc['Versicolor','petal.length']
dfSL3 = df.loc['Virginica','petal.length']
# print(dfSL1)
# print(dfSL1.dtype)

## 종류별로 그리기
axes[1,0].plot(dfSL1.index, dfSL1.values)
axes[1,0].plot(dfSL2.index, dfSL2.values)
axes[1,0].plot(dfSL3.index, dfSL3.values)

axes[1,0].set_title('petal.length')

###################################################
## 종류별로 petal.width 추출
dfSL1 = df.loc['Setosa','petal.width']
dfSL2 = df.loc['Versicolor','petal.width']
dfSL3 = df.loc['Virginica','petal.width']
# print(dfSL1)
# print(dfSL1.dtype)

## 종류별로  그리기
axes[1,1].plot(dfSL1.index, dfSL1.values)
axes[1,1].plot(dfSL2.index, dfSL2.values)
axes[1,1].plot(dfSL3.index, dfSL3.values)

axes[1,1].set_title('petal.width')

plt.show()
```

```python
## 산점도로 다시
df = pd.read_csv('../DATA/iris.csv')
# display(df.head())


setosa = df[ df['variety'] == 'Setosa' ]
versicolor = df[df['variety'] == 'Versicolor']
virginica = df[df['variety'] == 'Virginica']

plt.figure()

# Setosa
plt.scatter(setosa['petal.length'], setosa['petal.width'], 
            label='Setosa', 
            color='b', 
            marker='o')

# Versicolor
plt.scatter(versicolor['petal.length'], versicolor['petal.width'], 
            label='Versicolor', 
            color='orange', 
            marker='x')

# Virginica
plt.scatter(virginica['petal.length'], virginica['petal.width'], 
            label='Virginica', 
            color='g', 
            marker='^')

plt.title('품종 분류 시각화 (Petal Length vs Petal Width)')
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.legend()

# 결과 출력
plt.show()
```
