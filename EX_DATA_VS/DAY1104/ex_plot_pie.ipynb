### 【  Matplotlib의 Pie 그래프 】
- 전체 대비 항목에 대한 비중/비율 시각화 그래프
- 데이터 제공 시 자동 계산 수행

```python
## 모듈 로딩
import matplotlib.pyplot as plt
import koreanize_matplotlib
```

```python
## 데이터 준비
labels = 'Frogs', 'Hogs', 'Dogs', 'Logs'
sizes = [15, 30, 45, 10]

## 시각화 [기본]
plt.pie(sizes, labels=labels)
plt.show()

# wedges, texts=plt.pie(sizes, labels=labels, counterclock=False)
# #plt.setp(autotexts, size=8, weight="bold")
# print(texts, sep='\n\n')
# plt.setp(texts, fontsize=30)
# plt.show()
```

```python
## ------------------------------------------------------
## 조각별 비율 출력 => autopct 매개변수 : 00.00%
## ------------------------------------------------------
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.show()
```

```python
## ------------------------------------------------------
## 조각별 색상 설정 => colors 매개변수 : 색상리스트 
## ------------------------------------------------------
plt.pie(sizes, labels=labels, 
        autopct='%1.1f%%', 
        colors=["#802020", 'lightgreen', "#1B9CD8", 'yellow'])

plt.title("[항목별 비율]")
plt.show()
```

```python
## -----------------------------------------------------------
## 라벨/퍼센트값 위치 조절 => pctdistance/labeldistance 매개변수
##                         [기 0.6]       [기 1.1]
## -----------------------------------------------------------
plt.pie(sizes, labels=labels, 
        autopct='%1.1f%%', 
        colors=["#802020", 'lightgreen', "#1B9CD8", 'yellow'],
        pctdistance=1.2, labeldistance=0.6)

plt.title("[항목별 비율]")
plt.show()
```

```python
## -----------------------------------------------------------
## 조각 중심점에 떨어지는 간격 => explode 매개변수
## 그림자 효과 설정          => shadow 매개변수
## -----------------------------------------------------------
plt.pie(sizes, labels=labels, 
        autopct='%1.1f%%', 
        colors=["#802020", "#ECECEC", "#ECECEC", '#ECECEC'],
        pctdistance=1.2, labeldistance=0.6, 
        shadow=True,
        explode=[0.2, 0, 0, 0])

plt.title("[항목별 비율]")
plt.show()
```

```python
## -----------------------------------------------------------
## 시작 각도 설정           => startangle = 0 [기]
## 방향                    => couterclock = True [기]
## -----------------------------------------------------------
plt.pie(sizes, labels=labels, 
        autopct='%1.1f%%', 
        colors=["#802020", "#ECECEC", "#ECECEC", '#ECECEC'],
        pctdistance=1.2, labeldistance=0.6, 
        shadow=True,
        explode=[0.2, 0, 0, 0],
        startangle=180,
        counterclock=False)

plt.title("[항목별 비율]")
plt.show()
```

```python
## -----------------------------------------------------------
## 도넛 그래프 : 파이 그래프 변형
## - 쐐기 wedgeprops 매개변수 설정 : dict 타입
## -----------------------------------------------------------
wedgeDict = {'width':0.7, 
             'linewidth':5, 'edgecolor': 'white'}
plt.pie(sizes, labels=labels, 
        autopct='%1.1f%%', 
        colors=["#802020", "#FCA504D8", "#EFFB0D", '#ECECEC'],
        pctdistance=1.2, 
        labeldistance=0.6,
        wedgeprops=wedgeDict, 
        explode=[0.2, 0, 0, 0])

plt.title("[항목별 비율]")
plt.show()
```

```python
## -----------------------------------------------------------
## 범례 : pie(labels)값을 항목으로 표시
## -> loc='best'  : 'upper left', 'upper right', 'lower left'..
## -----------------------------------------------------------
wedgeDict = {'width':0.7, 'linewidth':5, 'edgecolor': 'white'}
plt.pie(sizes, labels=labels, 
        autopct='%1.1f%%', 
        colors=["#802020", "#FCA504D8", "#EFFB0D", '#ECECEC'],
        pctdistance=1.2, 
        labeldistance=0.6,
        wedgeprops=wedgeDict, 
        explode=[0.2, 0, 0, 0])

## 현재 그래프의 figure, axes
fig  = plt.gcf()
fig.set_facecolor('lightyellow')

## 공통 요소
plt.legend(loc='upper right', bbox_to_anchor=(1.3, 1.2))
plt.title("[항목별 비율]")
plt.show()
```
